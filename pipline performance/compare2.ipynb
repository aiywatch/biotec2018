```python
import pandas as pd
```

```python
data1 = pd.read_csv("data/pipeline.csv", index_col=0)
data2 = pd.read_csv("data/ncbi.csv", index_col=0).reset_index(drop=True)
data3 = pd.read_csv("data/qiime.csv", index_col=0)
```

```python
TAX = ["species", "genus", "family", "order", "class", "phylum", "superkingdom"]
```

```python
print(data1["count"].sum())
data1.head()
```

```python
print(data2["count"].sum())
data2.head()
```

```python
print(data3.shape)
data3.head()
```

# each rank count

```python
print("Pipeline : ")
for rank in TAX:
    count = data1[data1[rank].notnull()]["count"].sum()
    print(rank, count)
```

```python
print("NCBI : ")
for rank in TAX:
    count = data2[data2[rank].notnull()]["count"].sum()
    print(rank, count)
```

```python
print("QIIME 2 : ")
data3.info()
```
